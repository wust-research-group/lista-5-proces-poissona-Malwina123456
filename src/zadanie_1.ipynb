```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poisson(T, lam):
    t = [0]
    S = 0

    while True:
        u = np.random.uniform()
        S = S - np.log(u) * (1 / lam)
        if S > T:
            break
        t.append(S)

    return t
```

```python
def plot_poisson(T, lambdy):
    plt.figure(figsize=(12, 6))

    for lam in lambdy:
        t = poisson(T, lam)
        skoki = list(range(len(t)))  # range na listÄ™
        plt.step(t + [T], skoki + [skoki[-1]], label=f"lambda = {lam}")

    plt.title(f"Trajektorie procesu Poissona")
    plt.xlabel("Czas")
    plt.ylabel("Proces Poissona")
    plt.legend()
    plt.grid(True)
    plt.show()
```
